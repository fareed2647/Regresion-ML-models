# Sending data to Machine Learning

```python
from sklearn.model_selection import train_test_split
```

```python
Xtrain,xtest,Ytrain,ytest = train_test_split(Input,output,test_size= 0.25,random_state=32)
```

Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
L_regression = LinearRegression()
```

```python
L_regression.fit(Xtrain,Ytrain)
```

```python
y_train_pred = L_regression.predict(Xtrain)
```

```python
y_test_pred = L_regression.predict(xtest)
```

```python
from sklearn.metrics import r2_score
```

```python
Train_accuracy = r2_score(Ytrain,y_train_pred)
Test_accuracy =  r2_score(ytest, y_test_pred)
```

```python
print("TRAIN ACCURACY : ", Train_accuracy)
print("TEST ACCURACY : ", Test_accuracy)
```

Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
D_regressor = DecisionTreeRegressor(max_depth=6,min_samples_split=4,random_state=32 )
```

```python
D_regressor.fit(Xtrain,Ytrain)
```

```python
y_train_pred = D_regressor.predict(Xtrain)
y_test_pred = D_regressor.predict(xtest)
```

Train_accuracy = r2_score(Ytrain,y_train_pred)
Test_accuracy =  r2_score(ytest, y_test_pred)

```python
print("TRAIN ACCURACY : ", Train_accuracy)
print("TEST ACCURACY : ", Test_accuracy)
```

Random Forest through GridSearchCV

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
```

```python
RF_regressor = RandomForestRegressor()
```

```python
RF_regressor.fit(Xtrain,Ytrain)
```

```python
para = {'n_estimators':[80,100,120,140],
       'max_depth': [14,12,10,8],
       'min_samples_split': [2,3,4]}
```

```python
Grid = GridSearchCV(estimator=RF_regressor,param_grid = para,cv=5)
```

```python
Grid.fit(Xtrain,Ytrain)
```

```python
Grid.best_params_
```

```python
y_train_pred = Grid.predict(Xtrain)
y_test_pred = Grid.predict(xtest)
```

```python
Train_accuracy = r2_score(Ytrain,y_train_pred)
Test_accuracy =  r2_score(ytest, y_test_pred)
```

```python
print("TRAIN ACCURACY : ", Train_accuracy)
print("TEST ACCURACY : ", Test_accuracy)
```
